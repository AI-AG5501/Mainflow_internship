```python
import matplotlib.pyplot as plt #importing matlplotlib library.
import pandas as pd #importing pandas library.
import numpy as np #importing numpy library.
import seaborn as sns #importing seaborn library.
```

```python
df_frd=pd.read_csv('HouseData.csv') #loading csv file to the notebook.
```

```python
df_frd.head()
```

```python
#plotting scatter plot.
plt.scatter(df_frd['SaleCondition'],df_frd['SalePrice'])

#setting label for the plot.
plt.title('Scatter Plot') 

#setting label for x axis
plt.xlabel('SaleCondition')

#setting label for y axis.
plt.ylabel('SalePrice')

plt.show()
```

```python
#Line chart 

plt.plot(df_frd['YrSold'])
plt.plot(df_frd['SalePrice'])

#Adding title to the line chart.
plt.title('Line chart')

#setting the x and y labels.
plt.xlabel('YrSold')
plt.ylabel('SalePrice')

plt.show()
```

```python
#Bar chart.
plt.bar(df_frd['YrSold'],df_frd['SalePrice'])

#Adding title to the plot.
plt.title('Bar chart')

#setting x and y labels.
plt.xlabel('YrSold')
plt.ylabel('SalePrice')

plt.show()
```

```python
#Histogram

plt.hist(df_frd['SalePrice'])

#Adding title.
plt.title('Histogram')

plt.show()
```
